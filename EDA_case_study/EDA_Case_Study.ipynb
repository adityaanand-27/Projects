```python
def stack(var1):
    data = df_merge.pivot_table(values='TARGET', index=var1,columns='NAME_CONTRACT_STATUS',aggfunc='count')
    data=data.div(data.sum(axis=1),axis='rows')*100
    data.plot(kind='bar',stacked=True,figsize=(14,6))
    plt.title('Effect Of {0} on Loan Approval'.format(var1))
    plt.xlabel(var1)
    plt.ylabel('NAME_CONTRACT_STATUS %')
    plt.show()   
```
